## loading data from csv file

```python
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
netflix_df.head()
```

### filter data by movie type only

```python
netflix_df_movies_only=netflix_df[netflix_df["type"]=="Movie"]
```

```python
netflix_df_movies_only
```

### filter columns

```python
netflix_movies_col_subset=netflix_df_movies_only[["title", "country", "genre", "release_year","duration"]]
```

```python
netflix_movies_col_subset
```

```python
plt.scatter(netflix_movies_col_subset["release_year"],netflix_movies_col_subset["duration"])
```

```python
fig = plt.figure(figsize=(12,8))#increase the size of the plot 

plt.scatter(netflix_movies_col_subset["release_year"],netflix_movies_col_subset["duration"])
plt.title("Movie Duration by Year of Release")
plt.show()
```

### some movies are under 60.. lets findout the genre of those

```python
netflix_movies_col_subset[netflix_movies_col_subset["genre"]]
```

```python
short_movies=netflix_movies_col_subset[netflix_movies_col_subset["duration"]<60]
```

```python
short_movies.head(20)
```

#### most of them are Documentaries,Stand-Up,Children	lets visualize them with various colors

```python
colors=[]
for label,row in netflix_movies_col_subset.iterrows():
    if row["genre"]=="Children":
        colors.append("red")
    elif row["genre"]=="Documentaries":
        colors.append("blue")
    elif row["genre"]=="Stand-Up":
        colors.append("green")
    else:
        colors.append("black")
```

```python
colors[:10]
```

```python
fig = plt.figure(figsize=(12,8))#increase the size of the plot 

plt.scatter(netflix_movies_col_subset["release_year"],netflix_movies_col_subset["duration"],c=colors)
plt.title("Movie Duration by Year of Release")
plt.show()
```

```python
plt.style.use('fivethirtyeight')

fig = plt.figure(figsize=(12,8))#increase the size of the plot 

plt.scatter(netflix_movies_col_subset["release_year"],netflix_movies_col_subset["duration"],c=colors)
plt.show()
```
